```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

## CNN

```python
train_df = pd.read_csv("./CNN/train_res.csv")
val_df = pd.read_csv("./CNN/val_res.csv")
x = range(1,len(train_df)+1) 
plt.plot(x, train_df["acc"], label="train")
plt.plot(x, val_df["acc"], label="val")
plt.legend()
plt.title("CNN Accuracy of training and validation sets")
plt.xlabel("epoch")
plt.ylabel("Acc")
plt.show()
```

```python
train_df = pd.read_csv("./CNN/train_res.csv")
val_df = pd.read_csv("./CNN/val_res.csv")
x = range(1,len(train_df)+1) 
plt.plot(x, train_df["loss"], label="train")
plt.plot(x, val_df["loss"], label="val")
plt.legend()
plt.title("CNN Loss of training and validation sets")
plt.xlabel("epoch")
plt.ylabel("Loss")
plt.show()
```

## tofu+CNN

```python
train_df = pd.read_csv("./tofu+CNN/source_train_res.csv")
val_df = pd.read_csv("./tofu+CNN/source_val_res.csv")
x = range(1,len(train_df)+1) 
plt.plot(x, train_df["avg_acc"], label="train_avg")
plt.plot(x, train_df["worst_acc"], label="train_worst")
plt.plot(x, val_df["acc"], label="val")
plt.legend()
plt.title("tofu+CNN Source Task \n Accuracy of training and validation sets")
plt.xlabel("epoch")
plt.ylabel("Acc")
plt.show()
```
